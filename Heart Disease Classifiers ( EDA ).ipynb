```python
import sklearn.datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
readdata=pd.read_csv(r"C:\Users\adhit\ML Data sets\Inhouse project\heart.csv")
data = pd.DataFrame(readdata)
#data = data.drop([87,166,192,266,287,302],axis=0)
print(data.head())
print(data.shape)
print(data['target'].value_counts())
```

```python
count = 0
ages = []

for i in data['target']:
    count += 1
    if i == 1:
        ages.append(data['age'][count])
        
print(ages)        
```

```python
print(min(ages))
```

```python
sns.distplot(ages, kde = False, color ='red', bins = 30, axlabel = 'Age')
```

```python
sns.distplot(ages, rug=True, rug_kws={"color": "r"},
                  kde_kws={"color": "k", "lw": 3, "label": "KDE"},
                  hist_kws={"histtype": "step", "linewidth": 3,
                            "alpha": 1, "color": "g"})
```

```python
import matplotlib.pyplot as plt
classifiers = ['Support Vector (kernel = rbf)', 'Support Vector (kernel = linear)', 'Logistic Regression', 'Multi-Linear Preceptron',
        'Naive Bayes', 'K-Nearest Neighbors', 'Gradient Boost', 'XGBoost', 'Decision Tree', 'Random Forest']
#classifiers = ['Support Vector (kernel = rbf)', 'Naive Bayes', 'Logistic Regression', 'Multi-Linear Preceptron',  
               #'K-Nearest Neighbors', 'Gradient Boost', 'Random Forest', 'Decision Tree', ]
# getting values against each value of y
#students = [0.8390, 0.9609, 0.8390, 0.7170, 0.8292, 1.0, 1.0, 1.0, 0.9756, 0.8585]
accuracies = [0.7207, 0.8116, 0.8181, 0.8376, 0.8506, 0.9123, 0.9415, 0.9902, 1.0, 1.0]
plt.barh(classifiers, accuracies, color=['black', 'red', 'green', 'blue', 'brown', 'grey', 'pink', 'crimson', 'cyan', 'maroon'])
 
# setting label of y-axis
plt.ylabel("Classifiers")
 
# setting label of x-axis
plt.xlabel("Accuracy")
plt.title("Classifiers and their accuracies")
plt.show()
```
